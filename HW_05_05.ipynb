```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris_dataset.csv")
print(df.head())
```

```python
x = df[['sepal length (cm)', 'sepal width (cm)']]
y = df['target']
```

```python
Ptest = x.iloc[148]
Ptest_label = y.iloc[148]

x_train = x.drop(148)
y_train = y.drop(148)
```

```python
distances = np.sqrt(((x_train - Ptest) ** 2).sum(axis=1))

nearest_index = distances.idxmin()
predicted_target = y_train[nearest_index]

print("=== k1 mod ===")
print("Real class is: ", Ptest_label)
print("Predicted class is: ", predicted_target)
```

```python
k = 3
nearest_index = distances.nsmallest(k).index
nearest_target = y_train[nearest_index]

predicted_target = nearest_target.mode()

print("=== k3 mod ===")
print("Predicted class is: ", predicted_target)
```
