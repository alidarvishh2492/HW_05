```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris_dataset.csv")
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
x_train = x[:-1]
y_train = y[:-1]

x_test = x[-1]
y_test = y[-1]
```

```python
k = 3
distances = []

for i in range(len(x_train)):
    diff = x_test - x_train[i]
    d = np.sqrt(np.sum(diff ** 2))
    distances.append([d, y_train[i]])
```

```python
distances = sorted(distances, key=lambda x: x[0])
```

```python
neighbors_target = [distances[0][1], distances[1][1], distances[2][1]]

predicted_target = max(set(neighbors_target), key=neighbors_target.count)
correct_predicted = (predicted_target == y_test)

print("=== Non-normalized mode ===")
print("Real class is: ", y_test)
print("Predicted class is: ", predicted_target)
```

```python
mu = x_train.mean()
sigma = x_train.std()

x_train_normal = (x_train - mu) / sigma
x_test_normal = (x_test - mu) / sigma
```

```python
distances_normal = []

for i in range(len(x_train_normal)):
    diffrent_normal = x_test_normal - x_train_normal[i]
    d_normal = np.sqrt(np.sum(diffrent_normal ** 2))
    distances_normal.append([d_normal, y_train[i]])

distances_norm = sorted(distances_normal, key=lambda x: x[0])

neighbors_target_normal = [
    distances_normal[0][1],
    distances_normal[1][1],
    distances_normal[2][1],
]
```

```python
predicted_target_normal = max(set(neighbors_target_normal), key=neighbors_target_normal.count)
correct_predicted_normal = (predicted_target == y_test)

print("=== normalized mode ===")
print("Real class is: ", y_test)
print("Predicted class is: ", predicted_target)
```

```python
print("=== Final comparison ===")
print("Non-normalized mode is :", correct_predicted)
print("normalized mode is ", correct_predicted_normal)
```
